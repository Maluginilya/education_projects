```python
import pandas as pd
import numpy as np
pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.2f}'.format)
import seaborn as sns
```

```python
%%time
status_vs_tt = pd.read_csv('status_vs_tt.csv', sep = ',') # таблица Макса по выручкам
status_vs_tt = status_vs_tt.rename(columns = {'id_tt' : 'id_tt_cl'})

zao_info = pd.read_csv('zao_info.csv', sep = ';') # инфа о магазинах

febr = pd.read_csv('1_febr.csv', sep = ';')
march = pd.read_csv('1_march.csv', sep = ';')
april = pd.read_csv('1_april.csv', sep = ';')
may = pd.read_csv('1_may.csv', sep = ';')

ost_1 = pd.read_csv('1_ost.csv', sep = ';') # инфа о продуктах и остатках
```

```python
febr_1 = febr \
    .query('order_type == "offline"') \
    .groupby('id_tt_cl') \
    .agg({'BaseSum' : 'sum'}) \
    .rename(columns = {'BaseSum' : 'sum_feb'})
#febr_1
```

```python
march_1 = march \
    .query('order_type == "offline"') \
    .groupby('id_tt_cl') \
    .agg({'BaseSum' : 'sum'}) \
    .rename(columns = {'BaseSum' : 'sum_march'})
#march_1
```

```python
april_1 = april \
    .query('order_type == "offline"') \
    .groupby('id_tt_cl') \
    .agg({'BaseSum' : 'sum'}) \
    .rename(columns = {'BaseSum' : 'sum_april'})
#april_1
```

```python
may_1 = may \
    .query('order_type == "offline"') \
    .groupby('id_tt_cl') \
    .agg({'BaseSum' : 'sum'}) \
    .rename(columns = {'BaseSum' : 'sum_may'})
#may_1
```

```python
%%time
end = febr_1.merge(march_1, how = 'left', on = 'id_tt_cl').reset_index()
end_2 = april_1.merge(may_1, how = 'left', on = 'id_tt_cl').reset_index()
end_3 = end.merge(end_2, how = 'left', on = 'id_tt_cl').reset_index()
```

```python
end_shops = zao_info_end.merge(end_3, how = 'left', on = 'id_tt_cl').reset_index()
```

```python
def magazine_status(value):
    if value <= 3:
        return 'junior'
    elif 4 <= value <= 6:
        return 'middle'
    else:
        return 'senior'
```

```python
%%time
end_shops['magazine_status'] = end_shops.apply(lambda x: magazine_status(x['shtat']), axis = 1)
```

```python
zao_count = end_shops.groupby('magazine_status', as_index = False) \
    .agg({'id_tt_cl' : 'nunique'})
zao_count['%'] = (zao_count['id_tt_cl'] / 84).mul(100).round(2)
```

```python
zao_count = zao_count.rename(columns = {'magazine_status' : 'Класс магазина', 'id_tt_cl' : 'Количество точек'})
zao_count
```

```python
end_shops = end_shops.merge(status_vs_tt, how = 'left', on = 'id_tt_cl')
```

```python
end_shops = end_shops.drop(['level_0', 'index', 'Unnamed: 0'], axis = 1)
```

```python
end_shops['qty_kassa'] = end_shops['qty_kassa'] - end_shops['qty_kassa_so']
end_shops.rename(columns = {'qty_kassa' : 'off_cashbox', 'qty_kassa_so' : 'so_cashbox'})
```

```python
end_shops
```

```python
kassi_count = end_shops.groupby('magazine_status') \
    .agg({'qty_kassa' : 'sum', 'qty_kassa_so' : 'sum'})
```

```python
kassi_count['kass_procent'] = (kassi_count['qty_kassa'] / (kassi_count['qty_kassa'] + kassi_count['qty_kassa_so'])).mul(100).round(2)
kassi_count['kass__so_procent'] = (kassi_count['qty_kassa_so'] / (kassi_count['qty_kassa'] + kassi_count['qty_kassa_so'])).mul(100).round(2)
kassi_count
```

```python
kassi_count = kassi_count.rename(columns = {'qty_kassa' : 'Кол-во офф. касс', 'qty_kassa_so' : 'Кол-во касс СО', 'kass_procent' : '% офф. касс', 'kass__so_procent' : '% касс СО'})
kassi_count
```

```python
shop_month_perc = shops_month.copy()
shop_month_perc['sum_feb'] = round((shop_month_perc['sum_feb'] / fe) * 100, 2)
shop_month_perc['sum_march'] = round((shop_month_perc['sum_march'] / mar) * 100, 2)
shop_month_perc['sum_april'] = round((shop_month_perc['sum_april'] / ap) * 100, 2)
shop_month_perc['sum_may'] = round((shop_month_perc['sum_may'] / ma) * 100, 2)
shop_month_perc
```

```python
shops_month_for_one = shops_month.copy()
shops_month_for_one
```

```python
shops_month_for_one.sum_feb = shops_month.sum_feb / zao_count['Количество точек']
shops_month_for_one.sum_march = shops_month.sum_march / zao_count['Количество точек']
shops_month_for_one.sum_april = shops_month.sum_april / zao_count['Количество точек']
shops_month_for_one.sum_may = shops_month.sum_may / zao_count['Количество точек']
shops_month_for_one
```

```python
end_mid = end_shops.query('magazine_status == "middle"')
end_mid['deviation_feb'] = end_shops['sum_feb'] - shops_month_for_one.loc[1][1]
end_mid['deviation_march'] = end_shops['sum_march'] - shops_month_for_one.loc[1][2]
end_mid['deviation_april'] = end_shops['sum_april'] - shops_month_for_one.loc[1][3]
end_mid['deviation_may'] = end_shops['sum_may'] - shops_month_for_one.loc[1][4]
```

```python
end_sen = end_shops.query('magazine_status == "senior"')
end_sen['deviation_feb'] = end_shops['sum_feb'] - shops_month_for_one.loc[2][1]
end_sen['deviation_march'] = end_shops['sum_march'] - shops_month_for_one.loc[2][2]
end_sen['deviation_april'] = end_shops['sum_april'] - shops_month_for_one.loc[2][3]
end_sen['deviation_may'] = end_shops['sum_may'] - shops_month_for_one.loc[2][4]
end_sen
```

```python
shop_stat = pd.concat([end_mid, end_sen], axis=0)
shop_stat = shop_stat.query('format == "ВкусВилл"')
```

```python
shop_stat_testing = shop_stat[['id_tt_cl', 'ploshad', 'shtat','magazine_status', 'deviation_feb', 'deviation_march', 'deviation_april', 'deviation_may']]
shop_stat_testing['id_tt_cl'] = shop_stat_testing['id_tt_cl'].astype(object)
```

```python
shop_stat_mid = shop_stat_testing.query('magazine_status == "middle"').reset_index().drop(['index'], axis = 1)
shop_stat_mid['ploshad'] = shop_stat_mid['ploshad'].astype(object)
shop_stat_mid['shtat'] = shop_stat_mid['shtat'].astype(object)
shop_stat_mid = shop_stat_mid.sort_values(['shtat', 'deviation_may'], ascending = False)
```

```python
shop_stat_sen = shop_stat_testing.query('magazine_status == "senior"').reset_index().drop(['index'], axis = 1)
shop_stat_sen['ploshad'] = shop_stat_sen['ploshad'].astype(object)
shop_stat_sen['shtat'] = shop_stat_sen['shtat'].astype(object)
shop_stat_sen = shop_stat_sen.sort_values(['shtat', 'deviation_may'], ascending = False)
```
